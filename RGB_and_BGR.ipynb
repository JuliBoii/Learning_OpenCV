```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((300,300, 3), dtype='uint8')
```

```python
plt.imshow(img)
```

```python
img[:,0:100, 0] = 255
img[:,100:200,1] = 255
img[:,200:300,2] = 255

plt.imshow(img)
```

```python
cv2.imshow('RGB',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
