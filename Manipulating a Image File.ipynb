```python
im2 = im.copy()
```

```python
im2[23:100] = 120
```

```python
im2[23:100,40:100] = 255
```

```python
plt.imshow(im2,cmap='gray')
```

```python
im2[300:400,40:100] = 0
```

```python
plt.imshow(im2, cmap='gray')
```

```python
plt.imsave(r'/Users/angelito/Documents/python_projects/Learn/Learning_OpenCV/imgs/eagle_modified.jpg', im2, cmap='gray')
```
